```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
data_description = df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['median_house_value'], kde=True, color='blue')
plt.title('DistribuciÃ³n del Valor Medio de las Viviendas')
plt.xlabel('Valor Medio de la Vivienda')
plt.ylabel('Frecuencia')
plt.show()
```

```python
numerical_data = df.select_dtypes(include=['float64', 'int64'])
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de CorrelaciÃ³n entre las Variables NumÃ©ricas')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='median_income', y='median_house_value', data=df, alpha=0.6)
plt.title('RelaciÃ³n entre Ingreso Medio y Valor Medio de la Vivienda')
plt.xlabel('Ingreso Medio')
plt.ylabel('Valor Medio de la Vivienda')
plt.show()
```

La grÃ¡fica de dispersiÃ³n muestra una clara tendencia de aumento del valor de la vivienda a medida que el ingreso medio aumenta. Esta relaciÃ³n parece bastante lineal, lo cual un modelo de regresiÃ³n podria funcionar.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='ocean_proximity', data=df, palette='viridis')
plt.title('DistribuciÃ³n de Proximidad al OcÃ©ano')
plt.xlabel('Proximidad al OcÃ©ano')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
