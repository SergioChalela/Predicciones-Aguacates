```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
kilometraje = np.random.randint(10000, 100000, 100)
precio = 30000 - 0.15 * kilometraje + np.random.normal(0, 2000, 100)

# Visualizacin
plt.figure(figsize=(8, 6))
sns.scatterplot(x=kilometraje, y=precio, label="Datos Reales")
sns.regplot(x=kilometraje, y=precio, scatter=False, label="Modelo", color="red")
plt.xlabel("Kilometraje")
plt.ylabel("Precio Predicho")
plt.title("Relacin entre Kilometraje y Precio")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Datos de las mtricas para cada modelo
data = {
    "Modelo": ["Linear Regression", "Decision Tree", "Random Forest", "Gradient Boosting", "KNN"],
    "MSE": [145983500, 34658420, 19994500, 44191470, 18919220],
    "MAE": [7661.27, 3043.86, 2354.22, 3838.30, 2368.78],
    "R2": [0.205, 0.811, 0.891, 0.759, 0.897]
}

# Crear un DataFrame
df_metrics = pd.DataFrame(data)

# Configurar subplots para grficos de barras
fig, ax = plt.subplots(3, 1, figsize=(10, 18), sharex=True)

# Colores para cada mtrica
metrics = ["MSE", "MAE", "R2"]
titles = ["Error Cuadrtico Medio (MSE)", "Error Absoluto Medio (MAE)", "Coeficiente de Determinacin (R2)"]

for i, metric in enumerate(metrics):
    ax[i].bar(df_metrics["Modelo"], df_metrics[metric], alpha=0.7)
    ax[i].set_title(titles[i], fontsize=14)
    ax[i].set_ylabel(metric, fontsize=12)
    ax[i].tick_params(axis='x', rotation=45)

plt.xlabel("Modelo", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Regenerar grficos de barras para mtricas de modelos
fig, ax = plt.subplots(3, 1, figsize=(10, 18), sharex=True)

for i, metric in enumerate(metrics):
    ax[i].bar(df_metrics["Modelo"], df_metrics[metric], alpha=0.7, color='skyblue')
    ax[i].set_title(titles[i], fontsize=14)
    ax[i].set_ylabel(metric, fontsize=12)
    ax[i].tick_params(axis='x', rotation=45)

plt.xlabel("Modelo", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Crear grficos de mtricas individuales para asegurar compatibilidad
for metric, title in zip(metrics, titles):
    plt.figure(figsize=(10, 6))
    plt.bar(df_metrics["Modelo"], df_metrics[metric], color='skyblue', alpha=0.7)
    plt.title(title, fontsize=14)
    plt.ylabel(metric, fontsize=12)
    plt.xlabel("Modelo", fontsize=14)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import joblib
import matplotlib.pyplot as plt

# Cargar el modelo final y los datos de prueba
model = joblib.load("../models/final_model.pkl")
test_data = pd.read_csv("../data/test.csv")

# Separar caractersticas (X) y target (y)
X_test = test_data.drop(columns=["Price"])
y_test = test_data["Price"]

# Generar predicciones
predictions = model.predict(X_test)

# Crear el grfico de comparacin
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Valores Reales', alpha=0.6)
plt.scatter(range(len(predictions)), predictions, color='red', label='Predicciones', alpha=0.6)
plt.title("Comparacin entre Precios Reales y Predichos")
plt.xlabel("ndice")
plt.ylabel("Precio")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_pred = model.predict(X_test)

# Crear un DataFrame para los valores reales y predicciones
results_df = pd.DataFrame({'Real_Price': y_test, 'Predicted_Price': y_pred})

# Grfico comparativo
plt.figure(figsize=(10, 6))
plt.scatter(range(len(results_df)), results_df['Real_Price'], color='blue', label='Valores Reales', alpha=0.5)
plt.scatter(range(len(results_df)), results_df['Predicted_Price'], color='red', label='Predicciones', alpha=0.5)
plt.title('Comparacin entre Precios Reales y Predichos')
plt.xlabel('ndice')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
